## Gradient boosting:

* Gradient boosting is a machine learning technique that combines multiple weak prediction models, typically decision trees, to create a stronger and more accurate model. It is a popular and powerful algorithm for both regression and classification tasks.
* The main idea behind this algorithm is to build models sequentially and these subsequent models try to reduce the errors of the previous model.

#### GB working:
1. Initialization: The algorithm starts with an initial prediction (usually the average value for regression or the most common class for classification) for all data points in the training set.

2. Iterative training: A weak prediction model (e.g., decision tree) is trained on the residuals or errors of the previous predictions. The model is trained to minimize the loss function (e.g., mean squared error for regression or cross-entropy for classification).

3. Prediction update: The weak model's predictions are multiplied by a learning rate and added to the previous predictions. This step allows the model to gradually improve its predictions by making smaller adjustments at each iteration.

4. Repeat steps 2 and 3: The process is repeated by training additional weak models on the updated residuals and updating the predictions until a stopping criterion is met, such as reaching a maximum number of iterations or achieving satisfactory performance.

5. Final prediction: The final prediction is obtained by summing the predictions from all the weak models.
